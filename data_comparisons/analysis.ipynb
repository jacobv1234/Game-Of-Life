### Analysis of Game of Life results

```python
# module setup
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# convert numpy arrays to pandas dataframes
def createDataFrame(arr: np.ndarray, name: str) -> pd.DataFrame:
    return pd.DataFrame({
        "stabilised": arr[0],
        "population": arr[1],
        "name": name
    })
```

```python
# load numpy arrays
standard = createDataFrame(np.load('standard.npy'), "Conway's Game of Life")
highlife = createDataFrame(np.load('highlife.npy'), "HighLife")
bayshex = createDataFrame(np.load('bayhex.npy'), "Bays' Hexagonal Life")
simplehex = createDataFrame(np.load('simplehex.npy'), "Simple Hexagonal Life")

dataset = pd.concat([standard,highlife,bayshex,simplehex], ignore_index=True)
dataset.sample(5)
```

```python
# create combined graph
plt.scatter(standard['stabilised'], standard['population'], s=4, label = standard['name'][0])
plt.scatter(highlife['stabilised'], highlife['population'], s=4, label = highlife['name'][0])
plt.scatter(bayshex['stabilised'], bayshex['population'], s=4, label = bayshex['name'][0])
plt.scatter(simplehex['stabilised'], simplehex['population'], s=4, label = simplehex['name'][0])

# labels
plt.legend()
plt.xlabel('Generation Stabilised')
plt.ylabel('Final population')
plt.title('Comparison of different rulesets (Log scale)')

# logarithmic scale
plt.xscale('log')
plt.yscale('log')

plt.show()
```
